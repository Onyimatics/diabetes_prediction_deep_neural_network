#**A Deep Learning Approach For Diabetes Prediction**

In this blog, I will explain my approach in predicting diabetes using deep learning. The dataset contains 668 medical records with 8 features including glucose levels, BMI, and blood pressure which are critical indicators of a patient's risk of diabetes.




##**Data Preprocessing**
The preprocessing began with data cleaning which requires a domain knowledge. Columns like BMI, Blood Pressure and few others were found missing in some rows. A function was implemented to replace the zeros with the median value. Futhermore, the features were standardised using StandarScaler, ensuring that all inputs to the model share a common scale. This process is crucial for the efficient trainning of neural networks.


##**Model**

For the model architecture, a simplified sequential neural network was built. Initially, the network started with an input layer followed by two dense layers, with 128 and 64 neurons respectively. Though the accuracy was good, however, the model was significantly overfitted. This called for optimization.  

The optimization process used Keras Tuner's Hyperband algorithm to systematically search for otimal neural network hyperparameters including layer architecture, units, activationn functions, dropout rates, learning rate, and regularization strengths. This was followed by training the model with early stopping to maximise validation accuracy for diabetes prediction. Below is the optimised architectural design for the neural network that was built.

image.png

The HeUniform initializer was applied to optimize weight initialization, and L1/L2 regularization was incorporated to prevent overfitting. Batch normalization layers are added after each dense layer to stabilize and accelerate training, while dropout layers with rates of 20% and 30% help to further reduce the risk of overfitting by randomly deactivating neurons during training. The output layer used a sigmoid activation function appropriate for binary classification and traiing utilised 100 epochs.

##**Result**

At epoch 22, the best validation accuracy of approximately 90% was achieved and early stopping callback guided the model from overfitting and stopped training at epoch 32. More importantly, the model demonstrated good generalization capabilities, with minimal divergence between training and validation loss curves. The confusion matrix revealed balanced performance across both classes despite the original class imbalance in the dataset.
